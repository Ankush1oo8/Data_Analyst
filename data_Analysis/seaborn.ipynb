Seaborn is a Python data visualization library based on Matplotlib. It provides a high-level interface for creating informative and attractive statistical graphics.

```python
pip install seaborn
```

```python
import seaborn as sns
```

```python
##basic plotting
tips=sns.load_dataset('tips')
tips
```

```python
#create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of total bill vs tips")
```

```python
sns.lineplot(x='time',y='total_bill',data=tips)
plt.title("Line plot of total bill vs time")
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar plot of day vs total bill")
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
plt.title("Bar plot of sex vs total bill")
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box plot of day vs total bill")
```

```python
# violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("Violin plot of day vs total bill")
```

```python
## histogram
sns.histplot(tips['total_bill'],bins=10,kde=True)
plt.title("Histogram of total bill")
```

```python
#KDE PLOT
sns.kdeplot(tips['total_bill'],fill=True)
plt.title("KDE plot of total bill")
```

```python
##PAIR plot
sns.pairplot(tips)
plt.title("Pair plot of tips dataset")
```

```python
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation heatmap")
```
