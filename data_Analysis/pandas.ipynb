Pandas is a powerful and versatile Python library widely used for data analysis and manipulation. It's built on top of NumPy and provides efficient data structures like DataFrames and Series, along with a rich set of functions for data cleaning, transformation, and analysis.

```python
import pandas as pd
```

A Pandas Series is a one-dimensional labeled array capable of holding any data type (integers, strings, floating-point numbers, Python objects, etc.). The axis labels are collectively referred to as the index.



```python
data=[1,2,3,4,5,6]
series=pd.Series(data)
print(series)
```

```python
#create series from dict element
data={'a':1,'b':2,'c':3}
series=pd.Series(data)
print(series)
```

```python
data=[1,2,3,4,5,6]
series=pd.Series(data,index=['a','b','c','d','e','f'])
print(series)
```

```python
#dataframe
#craete a dataframe from dict of list
data={
    'Name':['Ankush','Katarya','Gotya','ujj'],
    'Age':[21,22,23,24],
    'City':['Banglore','New York','Florida','America']
}

dataFrame=pd.DataFrame(data)
print(dataFrame)
```

```python
## create df from list of dict
data=[
    {'Name':'Ankush','Age':21,'City':'Banglore'},
    {'Name':'Katarya','Age':22,'City':'New York'},
    {'Name':'Gotya','Age':23,'City':'Florida'},
]

df=pd.DataFrame(data)
print(df)
```
