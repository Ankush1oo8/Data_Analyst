```python
##Renaming col
df=df.rename(columns={'Date':'Sales Date'})
df.head(5)
```

```python
## change data types
df['Val_New']=df['Value'].fillna(df['Value'].mean()).astype(int)
```

```python
df['Squareed_val']=df['Val_New'].apply(lambda x:x*2)
df.head(5)
```

```python
## merging and joing dataframes
data1 = {'id': [1, 2, 3, 4],
         'name': ['Alice', 'Bob', 'Charlie', 'David'],
         'age': [25, 30, 22, 28]}
df1 = pd.DataFrame(data1)

data2 = {'id': [2, 3, 4, 5],
         'city': ['New York', 'London', 'Paris', 'Tokyo'],
         'salary': [50000, 60000, 70000, 80000]}
df2 = pd.DataFrame(data2)

df1
```

```python
df2['age']=[50,68,56,43]
```
