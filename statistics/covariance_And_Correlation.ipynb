```python
import seaborn as sns
```

```python
df=sns.load_dataset('healthexp')
```

```python
df.head(5)
```

```python
numerical_df = df.select_dtypes(include=['number'])  # Select only numerical columns
numerical_df.cov()  # Calculate covariance on the numerical subset
```

```python
## correalation pearson
```

```python

numerical_df.corr(method='pearson')
```

```python
numerical_df.corr(method='spearman')
```

```python
numerical_df.corr(method='kendall')
```
